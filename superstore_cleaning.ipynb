```python
import pandas as pd

df = pd.read_csv('Sample - Superstore.csv', encoding='latin-1')
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nNull values:\n", df.isnull().sum())
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month Name'] = df['Order Date'].dt.strftime('%b')
df['Profit Margin'] = (df['Profit'] / df['Sales'] * 100).round(2)

print("New columns added. Sample:")
df[['Order Date','Year','Month','Profit Margin']].head()
```

```python
print("=== Sales by Region ===")
print(df.groupby('Region')['Sales'].sum().round(2))

print("\n=== Profit by Region ===")
print(df.groupby('Region')['Profit'].sum().round(2))

print("\n=== Profit by Category ===")
print(df.groupby('Category')['Profit'].sum().round(2))

print("\n=== Top 5 Sub-Categories by Sales ===")
print(df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head())
```

```python
df.to_csv('superstore_clean.csv', index=False,encoding='utf-8')
print("Clean file saved! Total rows:", len(df))
```

```python
print("Total Profit:", df['Profit'].sum().round(2))
print("Total Sales:", df['Sales'].sum().round(2))
print("Profit Margin:", (df['Profit'].sum() / df['Sales'].sum() * 100).round(2))
```
